```python
def get_hs_codes(code):

    my_key = "&key=34e40301bda77077e24c859c6c6c0b721ad73fc7"

    hs = 'HS10'

    my_key = "&key=34e40301bda77077e24c859c6c6c0b721ad73fc7"

    end_use = "hs?get=CTY_NAME,CON_VAL_YR,I_COMMODITY_SDESC"

    surl = "https://api.census.gov/data/timeseries/intltrade/imports/" + end_use 

    surl  = surl + my_key + "&time=" + "2024-12" + "&COMM_LVL=" + "HS10" + "&I_COMMODITY=" + code + "*"

    url = surl + "&CTY_CODE=" + '-'

    r = requests.get(url) 
    
    print(r)
    
    df = pd.DataFrame(r.json()[1:]) # This then converts it to a dataframe
    # Note that the first entry is the labels

    df.columns = r.json()[0]

    return df["I_COMMODITY"]
```

```python
df = pd.DataFrame()

for index, row in exemptions_41225_df.iterrows():

    foo = get_hs_codes(row['HS_Code'])

    foo_df = pd.DataFrame(foo)  # Ensure it's a DataFrame
    
    df = pd.concat([df, foo_df], ignore_index=True) 

    
```
